```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.replace({True:1,False:0})
```

```python
dataset
```

```python
dataset.columns
```

```python
independant= dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
```

```python
dependant=dataset['Profit']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor()
regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
Y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_decisiontree.sav"
pickle.dump(regressor,open(filename,'wb'))
```
